```python
from distutils.util import change_root
from os import truncate
from pathlib import Path
import numpy
import math
import csv
with open("../PyBank/Resources/budget_data.csv") as file:
    csvreader = csv.reader(file)
    header = next(csvreader)
    i=0
    total = 0
    prev_row = 0
    change = 0
    changes = []
    min_change = 0
    max_change = 0
    min_change_date = None
    max_change_date = None
    gr_decr = [None,math.inf]
    gr_incr = [None,-math.inf]
    for row in csvreader:

        if i != 0: 
            change = -(prev_row - float(row[1]))
            print(f"The change is {change}")
            print(f"Date is {row[0]}")
            changes.append(change)
        prev_row = float(row[1])

        i +=1
    
  
        if change < gr_decr[1]:
            gr_decr = (row[0], change)
        print(gr_decr)
        if change > gr_incr[1]:
            gr_incr = (row[0], change)
        print(gr_incr)


        total += float(row[1])
       
    lst_avg = round(numpy.average(changes), 2)
       

        

output = f"""
Financial Analysis
----------------------------
 Total Months: {i}
 Total: ${total}
 Average: ${lst_avg}
 Greatest Decrease: ${gr_decr}
 Greatest Increase: ${gr_incr}


"""

print(output)
with open("output.txt", "w") as file:
    file.write(output)
```
